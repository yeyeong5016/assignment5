```python
class Heap:
    def __init__(self, *args):  
        if len(args) != 0:      
            self.__A = args[0] # íŒŒë¼ë¯¸í„°ë¡œ ì˜¨ ë¦¬ìŠ¤íŠ¸
        else:
            self.__A = []      

    # [ì•Œê³ ë¦¬ì¦˜ 8-2] êµ¬í˜„: í™ì— ì›ì†Œ ì‚½ì…í•˜ê¸°(ì¬ê·€ ì•Œê³ ë¦¬ì¦˜ ë²„ì „)
    def insert(self, x):
        self.__A.append(x)
        self.__percolateUp(len(self.__A)- 1)

# ìŠ¤ë©°ì˜¤ë¥´ê¸°
    def __percolateUp(self, i:int):
        parent = ( i - 1 ) // 2 

        if i>0 and self.__A[i] > self.__A[parent]:
            self.__A[i], self.__A[parent] = self.__A[parent], self.__A[i]
            self.__percolateUp(parent)

    # [ì•Œê³ ë¦¬ì¦˜ 8-2] êµ¬í˜„: í™ì—ì„œ ì›ì†Œ ì‚­ì œí•˜ê¸°
    def deleteMax(self):
    # heap is in self.__A[0...len(self.__A)-1]
        if (not self.isEmpty()):
            max = self.__A[0]
            self.__A[0] = self.__A.pop()
            # *.pop(): ë¦¬ìŠ¤íŠ¸ì˜ ëì›ì†Œ ì‚­ì œ í›„ ì›ì†Œ ë¦¬í„´
            self.__percolateDown(0)
            return max
        else:
            return None
        
    def __percolateDown(self, i: int):
    # PercolateDown w/ self.__A[i] as the root
        lchild = 2 * i + 1  # left child
        rchild = 2 * i + 2  # right child
        maxchild = lchild
                
        if (lchild <= len(self.__A)-1):
            if (rchild <= len(self.__A)-1 and self.__A[maxchild] < self.__A[rchild]):
                maxchild = rchild  # index of larger child
            if self.__A[i] < self.__A[maxchild]:
                self.__A[i], self.__A[maxchild] = self.__A[maxchild], self.__A[i]
                self.__percolateDown(maxchild)
    # ìƒì„±                  
    def buildHeap(self):
        for i in range((len(self.__A) - 2) // 2, -1, -1):
            self.__percolateDown(i)

    def max(self):
        return self.__A[0]
    
    def isEmpty(self):
         return len(self.__A) == 0
    
    def clear(self):
         self.__A = []
        
    def size(self):
        return len(self.__A)
    

import csv
from datetime import datetime

# Heap í´ë˜ìŠ¤ëŠ” ì‚¬ìš©ìê°€ ì œê³µí•œ ê·¸ëŒ€ë¡œ ì‚¬ìš©í•œë‹¤ê³  ê°€ì •
# ìƒì¼ì„ ê¸°ì¤€ìœ¼ë¡œ í•˜ëŠ” ì‚¬ìš©ì ì •ì˜ ë¹„êµ ê°ì²´
class BirthdayEntry:
    def __init__(self, name, birthday_str):
        self.name = name
        self.birthday = datetime.strptime(birthday_str, "%Y%m%d")
        
    def __lt__(self, other):
        # ìµœì‹  ìƒì¼ì¼ìˆ˜ë¡ ìš°ì„ ìˆœìœ„ê°€ ë†’ë„ë¡ ë¹„êµ
        return self.birthday.timestamp() < other.birthday.timestamp()

    def __str__(self):
        return f"{self.name} - {self.birthday.date()}"

# CSV íŒŒì¼ì—ì„œ ë°ì´í„° ì½ê³  í™ì— ì €ì¥
birthday_list = []
with open('birthday.csv', newline='', encoding='cp949') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        name = row['ì´ë¦„'].strip()
        bday_str = row['ìƒë…„ì›”ì¼8ìë¦¬(ì˜ˆ.20040101)'].strip()

        # ìƒì¼ ìœ íš¨ì„± ê²€ì‚¬
        if not bday_str or len(bday_str) != 8 or not bday_str.isdigit():
            continue  # ë¹„ì–´ìˆê±°ë‚˜ ì˜ëª»ëœ í˜•ì‹ ìƒëµ

        try:
            entry = BirthdayEntry(name, bday_str)
            birthday_list.append(entry)
        except ValueError:
            continue  # ë‚ ì§œ íŒŒì‹± ì‹¤íŒ¨ ì‹œ ìƒëµ

# í™ ìƒì„± ë° ë¹Œë“œ
heap = Heap(birthday_list)
heap.buildHeap()

# ìƒì¼ì´ ëŠ¦ì€ ìˆœì„œë¡œ 10ëª… ì¶œë ¥
print("ìƒì¼ì´ ëŠ¦ì€ ìˆœì„œ Top 10:")
for _ in range(min(10, heap.size())):
    person = heap.deleteMax()
    print(person)
```

## ê³¼ì œ 6. LeetCode 703. Kth Largest Element in Stream

```python
import heapq

class KthLargest:

    def __init__(self, k: int, nums: list[int]):
        self.k = k
        self.min_heap = nums
        heapq.heapify(self.min_heap)

        # kê°œë§Œ ë‚¨ê¸°ê³  ë‚˜ë¨¸ì§€ëŠ” ì œê±°
        while len(self.min_heap) > k:
            heapq.heappop(self.min_heap)

    def add(self, val: int) -> int:
        heapq.heappush(self.min_heap, val)
        
        # kê°œ ìœ ì§€
        if len(self.min_heap) > self.k:
            heapq.heappop(self.min_heap)

        # kë²ˆì§¸ë¡œ í° ê°’ì€ heapì˜ ë£¨íŠ¸
        return self.min_heap[0]
```
